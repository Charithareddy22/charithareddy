```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.regressionplots import influence_plot
import statsmodels.formula.api as smf
import numpy as np
```

```python
import pandas as pd
from google.colab import files
uploaded  = files.upload()
file_name = list(uploaded.keys())[0]
civil = pd.read_csv(file_name)
```

```python
X = civil[['Building_Height']]
y = civil['Construction_Cost']
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y)
```

```python
intercept = model.intercept_
coefficient = model.coef_[0]
```

```python
print(f"Equation of the regression line: Construction Cost = {intercept:.2f} + ({coefficient:.2f} * Building Height)")
```

```python
y_pred = model.predict(X)
```

```python
from sklearn.metrics import r2_score, mean_squared_error # Import necessary functions

y_pred = model.predict(X)

r2 = r2_score(y, y_pred)
mse = mean_squared_error(y, y_pred)
```

```python
print(f"R-squared: {r2:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
```

```python
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred, color='red', label='Regression line')
plt.xlabel('Building Height')
plt.ylabel('Construction Cost')
plt.title('Simple Linear Regression: Building Height vs Construction Cost')
plt.legend()
plt.show()
```
