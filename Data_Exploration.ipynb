```python
import pandas as pd
from google.colab import files
uploaded  = files.upload()
file_name = list(uploaded.keys())[0]
civil = pd.read_csv(file_name)
```

```python
dependent_variable = 'Construction Cost'
independent_variables = [col for col in civil.columns if col != dependent_variable]
```

```python
missing_values = civil.isnull().sum()
print("Missing values in each column:\n", missing_values)
```

```python
summary_statistics = civil.describe()
print("Summary statistics:\n", summary_statistics)
```

```python
correlation_matrix = civil.corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
