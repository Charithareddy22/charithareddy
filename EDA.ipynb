```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv("/content/Day_13_Pharma_data.csv")
```

```python
print(df.isnull().sum())
```

```python
df = df.drop_duplicates()
```

```python
# Group sales by region
sales_by_region = df.groupby("Region")["Sales"].sum().reset_index()

# Plot bar chart
plt.figure(figsize=(8, 5))
sns.barplot(x="Region", y="Sales", data=sales_by_region, palette="Blues_r")
plt.title("Total Sales per Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
#Scatterplot
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Marketing_Spend", y="Sales", data=df, alpha=0.7, color="purple")
plt.title("Marketing Spend vs Sales")
plt.xlabel("Marketing Spend")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
#Boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(x="Age_Group", y="Effectiveness", data=df, palette="coolwarm")
plt.title("Drug Effectiveness across Age Groups")
plt.xlabel("Age Group")
plt.ylabel("Effectiveness")
plt.xticks(rotation=45)
plt.show()
```

```python
#LinePlot
plt.figure(figsize=(10, 6))
# Verify the column name 'Product' exists in your DataFrame
print(df.columns)
# If the column name is different, update it in the hue parameter
sns.lineplot(x="Trial_Period", y="Sales", data=df, marker="o", palette="Set1")
plt.title("Sales Trend Over Trial Periods")
plt.xlabel("Trial Period")
plt.ylabel("Sales")
plt.legend(title="Product")
plt.grid(True)
plt.show()
```

```python
#Heatmap
plt.figure(figsize=(7, 5))
sns.heatmap(df[["Sales", "Marketing_Spend", "Effectiveness"]].corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
